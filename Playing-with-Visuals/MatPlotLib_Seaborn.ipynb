#### Data Loading

2020 FIFA Dataset Found <a href="https://www.kaggle.com/stefanoleone992/fifa-21-complete-player-dataset?select=players_20.csv" target="_blank"> Here</a>.

```python
df_20 = pd.read_csv('data/players_20.csv')
df_20.head()
```

```python
country = ['United States', 'Canada', 'England', 'Brazil', 'Argentina']

df_country = df_20[df_20['nationality'].isin(country)]
```

## Barplots

```python
barplot = df_country.groupby(['nationality'], as_index=False).mean()[['nationality', 'overall']]
barplot
```

```python
# Using MatplotLib to generate Barplot

plt.figure(figsize=(8,4), tight_layout=True)
colors = sns.color_palette('Set2')

plt.bar(x=barplot['nationality'], height=barplot['overall'], color=colors[:5])
plt.xlabel('Nationality')
plt.ylabel('Average Rating')
plt.title('Barplot')
plt.show()
```

It's difficult to see any ratings differences between the average players from Canada, England or the US. Only the top players make it to the national teams so we will examine those further.

#### Examine top 20 players per country using SeaBorn

```python
# Preprocess top 20 averages per country
df_top_20 = pd.concat(
[df_country[df_country['nationality']=='Argentina'][:20],
df_country[df_country['nationality']=='Brazil'][:20],
df_country[df_country['nationality']=='England'][:20],
df_country[df_country['nationality']=='Canada'][:20],
df_country[df_country['nationality']=='United States'][:20]])

barplot = df_top_20.groupby(['nationality'], as_index=False).mean()[['nationality', 'overall']]
```

```python
plt.figure(figsize=(8,4), tight_layout=True)
ax=sns.barplot(x=barplot['nationality'], y=barplot['overall'],palette='Set2', ci=None)
ax.set(title='Barplot', xlabel='Nationality', ylabel='Average Rating')
plt.show()
```

Rating differences are now more visible between the top 20 players from each country but we can still examine further - such as total number of players in the top bracket leagues

#### Barplot by "N" Variables

```python
# Preprocess for # of players in top leagues

barplot = df_country[df_country['league_name'].isin(['Spain Primera Division', 'Italian Serie A', 'German 1. Bundesliga'])]
barplot = barplot.groupby(['nationality', 'league_name'], as_index=False).count()
```

```python
# Plot with Seaborn
plt.figure(figsize=(12,6), tight_layout=True)
ax = sns.barplot(x=barplot['nationality'], y=barplot['sofifa_id'], hue=barplot['league_name'], palette='Set2')
ax.set(title='N* of Players outside Domestic League', xlabel='Country', ylabel='Count')
ax.legend(title='League', title_fontsize='13', loc='upper right')
plt.show()
```

This mapping clearly shows the difference in the number of players in top leagues - specifically the disparity between Argentina and Brazil vs Canadian, English and US teams. This exemplifies why Brazil and Argentina are huge football nations

## Histograms

```python
# Set plot size and number of bins 
plt.figure(figsize=(10,6), tight_layout=True)
bins=[160,165,170, 175,180,185,190,195,200]

# Set up MatplotLib Setting
plt.hist(df_country['height_cm'], bins=bins, color=sns.color_palette('Set2')[2], linewidth=2)
plt.title('Histogram')
plt.xlabel('Height (cm)')
plt.ylabel('Count')

# Set up seaborn settings
ax = sns.histplot(data=df_country, x='height_cm', bins=bins, color=sns.color_palette('Set2')[2], linewidth=2)
ax.set(title='Histogram', xlabel='Height (cm)', ylabel='Count')

plt.show()
```

Average player's height is between 175-185cm and therefore doesn't seem important in football

## Boxplots

Because of the distribution above, we can check to see height distribution of players from different nations

```python
plt.figure(figsize=(10,6), tight_layout=True)
ax = sns.boxplot(data=df_country, x='nationality', y='height_cm', palette='Set2', linewidth=2.5)
ax.set(title='Boxplot', xlabel='', ylabel='Height (cm)')
plt.show()
```

This boxplot indicates that Brazil and Argentina don't have taller than average players but are still successful in competitions

Argentina actually presents with the shortest players among the 5 countries

## Scatterplots

Let's look at the players weights vs their heights

```python
plt.figure(figsize=(10,6), tight_layout=True)
ax = sns.scatterplot(data=df_country, x='height_cm', y='weight_kg', hue='nationality', palette='Set2', s=60)
ax.set(xlabel='Height (cm)', ylabel='Weight (kg)')
ax.legend(title='Country', title_fontsize=12)
plt.show()
```

Weight & height appear to follow a basic linear regression pattern from the above scatterplot

Let's take a look at the value that players bring to their respective countries. We can examine this using **pie charts**

```python
# Preprocess for my Favourite Team, top 10 players
pie = df_20[df_20['club_name']=='Real Madrid']
pie = pie.sort_values('value_eur', ascending=False)[['short_name', 'value_eur']]
pie = pie[:15]
```

```python
# Design Pie Chart Layout
colors = sns.color_palette('pastel')
explode = [0]*(len(pie)-2) +[0.2] +[0]

plt.figure(figsize=(10,6), tight_layout=True)
plt.pie(pie['value_eur'],
        labels=pie['short_name'],
        autopct='%.0f %%',
        explode=explode,
        pctdistance=0.7,
        colors=colors,
        shadow=True)
plt.title('Real Madrid', weight='bold')
plt.show()
```

We can see that my favourite player Ramos may not be the most highly valued player on Real Madrid, but he is sitting at 14th place on the team and not far from the top 10.

But how can we see his impact on his club or country?

```python
# Process data
piechart_2 = df_20[df_20['nationality']=='Spain']
piechart_2 = piechart_2[:15].sort_values('value_eur')[['short_name', 'value_eur']]
```

```python
# Design Subplots for side-by-side piecharts
fig, ax = plt.subplots(1,2, figsize=(12,5), tight_layout=True)
colors = sns.color_palette('Set2')
explode = [0]*(len(pie)-2) +[0.2] +[0]
explode_2 = [0] + [0.2] + [0]*(len(piechart_2)-2)

# Plot 1
ax[0].pie(pie['value_eur'],
          labels=pie['short_name'], 
          autopct='%.0f %%', 
          explode=explode, 
          pctdistance=0.8,
          colors=colors,
          shadow=True)
ax[0].set_title('Real Madrid',weight='bold')

# plot 2
ax[1].pie(piechart_2['value_eur'],
          labels=piechart_2['short_name'], 
          autopct='%.0f %%', 
          explode=explode_2, 
          pctdistance=0.8,
          colors=colors,
          shadow=True)
ax[1].set_title('Spain', weight='bold')
plt.show()
```

This comparison shows that Ramos is valued about equally (5%) between his National team (Spain) and in his club (Real Madrid)

## Line Plots

```python
# load previous years of data for preprocessing
df_21 = pd.read_csv('data/players_21.csv')
df_20 = pd.read_csv('data/players_20.csv')
df_19 = pd.read_csv('data/players_19.csv')
df_18 = pd.read_csv('data/players_18.csv')
df_17 = pd.read_csv('data/players_17.csv')
df_16 = pd.read_csv('data/players_16.csv')
df_15 = pd.read_csv('data/players_15.csv')

df_list = [df_15, df_16, df_17, df_18, df_19, df_20, df_21]
for i,dataframe in enumerate(df_list):
    dataframe['year'] = 2015 + i

players = ['L. Messi', 'Cristiano Ronaldo', 'K. De Bruyne', 'K. Mbappé', 'V. van Dijk']
lineplot = pd.concat(df_list)[['short_name', 'overall', 'year']]
lineplot = lineplot[lineplot['short_name'].isin(players)]
lineplot = lineplot.pivot(index='short_name', columns='year', values='overall').T
lineplot.reset_index(inplace=True)
lineplot = lineplot.set_index('year')
```

```python
# customize plot
plt.figure(figsize=(10,6), tight_layout=True)
plt.plot(lineplot,'o-', linewidth=2)
plt.xticks([2015,2016,2017,2018,2019,2020,2021])
plt.xlabel('Years')
plt.ylabel('Ratings')
plt.title('Ratings over the years')
plt.legend(title='Players', title_fontsize=13, labels=players)
plt.show()
```
